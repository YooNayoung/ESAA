### Step 1. Import the necessary libraries

```python
import pandas as pd 
%matplotlib notebook 
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame that should look like the one below.

```python
dict = {'first_name':['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'last_name':['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
        'age':[42, 52, 36, 24, 73],
        'female':[0, 1, 1, 0, 1],
        'preTestScore':[4, 24, 31, 2, 3],
        'postTestScore':[25, 94, 57, 62, 70]}
df = pd.DataFrame(dict)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.scatter(x='preTestScore', y='postTestScore', data=df, s='age')
axes.set_title('Scatterplot of preTestScore and postTestScore')
axes.set_xlabel('preTestScore')
axes.set_ylabel('postTestScore');
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.scatter(x='preTestScore', y='postTestScore', data=df, s=df['postTestScore']*4.5, c='female')
axes.set_title('Scatterplot of preTestScore and postTestScore')
axes.set_xlabel('preTestScore')
axes.set_ylabel('postTestScore');
```
